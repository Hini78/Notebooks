```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.metrics import roc_auc_score, accuracy_score
from sklearn.compose import ColumnTransformer
```

```python
X = data.drop('Churn', axis=1)
y = data['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
#          
for title in num_cols:
  mid = X_train[title].mean()
  deviation = X_train[title].std()
  X_train[title] = (X_train[title] - mid) / deviation
  X_test[title] = (X_test[title] - mid) / deviation # X_test    X_train
X_train.sample(5)
```

```python
#     one hot encoding   
onehot = OneHotEncoder(sparse_output=False, drop='first')
new_data = onehot.fit_transform(X_train[cat_cols])
encoded_data = pd.DataFrame(new_data, columns=onehot.get_feature_names_out(cat_cols), index=X_train.index)
encoded_data = encoded_data.merge(X_train[num_cols], how="inner",left_index=True, right_index=True)
encoded_data.sort_index(inplace=True)
y_train.sort_index(inplace=True)

onehot = OneHotEncoder(sparse_output=False, drop='first')
new_data = onehot.fit_transform(X_test[cat_cols])
encoded_data_test = pd.DataFrame(new_data, columns=onehot.get_feature_names_out(cat_cols), index=X_test.index)
encoded_data_test = encoded_data_test.merge(X_test[num_cols], how="inner",left_index=True, right_index=True)
encoded_data_test.sort_index(inplace=True)
y_test.sort_index(inplace=True)
```

```python
logregCV = LogisticRegressionCV(cv=5)
logregCV.fit(encoded_data, y_train)

y_pred = logregCV.predict(encoded_data_test)
roc_auc = roc_auc_score(y_test, y_pred)
print("Roc auc : %.4f" % roc_auc)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy : %.4f" % accuracy)
```

```python
col_transformer = ColumnTransformer(#     
    [
        ('num_processing', RobustScaler(), num_cols), #   ' ', ,     
        ('categorial_columns_procesing', OneHotEncoder(sparse_output=False, drop='first'), cat_cols)
    ]
)
log_reg_pipeline = make_pipeline(col_transformer, LogisticRegression())

X = data.drop('Churn', axis=1)
y = data['Churn']
X_pipe_train, X_pipe_test, y_pipe_train, y_pipe_test = train_test_split(X, y, train_size=0.8, random_state=42)

param_grid={
    'logisticregression__C':[100, 10, 1, 0.1, 0.01, 0.001]
}
pipe_grid = GridSearchCV(log_reg_pipeline, param_grid, scoring='roc_auc')
pipe_grid.fit(X_pipe_train,y_pipe_train)

best_pipe_model = pipe_grid.best_estimator_
roc_auc_score_test = best_pipe_model.score(X_pipe_test, y_pipe_test)
print("ROC AUC   :", roc_auc_score_test)
```
